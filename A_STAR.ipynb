```python
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 1, 'E': 4},
    'C': {'F': 2},
    'D': {'G': 2},
    'E': {'G': 1},
    'F': {'G': 5},
    'G': {}
}
heuristic = {
    'A': 7, 'B': 6, 'C': 4,
    'D': 2, 'E': 1, 'F': 3,
    'G': 0
}
```

```python
def a_star(graph, start, goal, heuristic):
    open_list = [start]       # daftar node yang akan dieksplor
    closed_list = []          # daftar node yang sudah dikunjungi

    g = {start: 0}            # biaya nyata dari start ke node tertentu
    parent = {start: None}    # untuk melacak jalur yang diambil

    while open_list:
        # 1. pilih node dengan f = g + h terkecil
        current = min(open_list, key=lambda node: g[node] + heuristic[node])

        # 2. jika node saat ini adalah goal â†’ selesai
        if current == goal:
            path = []
            while current is not None:
                path.append(current)
                current = parent[current]
            path.reverse()
            return path, g[goal]

        # 3. pindahkan current dari open ke closed
        open_list.remove(current)
        closed_list.append(current)

        # 4. periksa semua tetangga current
        for neighbor, cost in graph[current].items():
            if neighbor in closed_list:
                continue

            # 5. hitung biaya baru
            new_g = g[current] + cost

            # 6. jika neighbor belum pernah dikunjungi atau ditemukan jalur lebih murah
            if neighbor not in open_list or new_g < g.get(neighbor, float('inf')):
                parent[neighbor] = current
                g[neighbor] = new_g
                if neighbor not in open_list:
                    open_list.append(neighbor)

    return None, float('inf')  # jika tidak ada jalur
```

```python
a_star(graph,'A','G',heuristic)
```
